```python
import numpy as np
import matplotlib.pyplot as plt

axisX = np.arange(46)
axisY = np.array([0.0000,0.0175,0.0349,0.0523,0.0698,0.0872,0.1045,0.1219,0.1392,0.1564,0.1736,0.1908,0.2079,0.2250,0.2419,0.2588,0.2756,0.2924,0.3090,0.3256,0.3420,0.3584,0.3746,0.3907,0.4067,0.4226,0.4384,0.4540,0.4695,0.4848,0.5000,0.5150,0.5299,0.5446,0.5592,0.5736,0.5878,0.6018,0.6157,0.6293,0.6428,0.6561,0.6691,0.6820,0.6947,0.7071])

# plot of sin function
plt.plot(axisX, axisY)
plt.show()
```

```python
from google.colab import files

# random function data
randomY = np.random.random(46)

# plot of random function
plt.plot(axisX, randomY)
plt.show()

#Середня абсолютна похибка вимірює середні відмінності між прогнозованими і фактичними значеннями.
def mae(y_true, predictions):
    y_true, predictions = np.array(y_true), np.array(predictions)
    return np.mean(np.abs(y_true - predictions))

#повертає середнє значення квадратів кожної різниці між оціненим і справжнім значенням.
def mse(actual, predicted):
    return np.square(np.subtract(np.array(actual), np.array(predicted))).mean()

mae = mae(axisY, randomY)
mse = mse(axisY, randomY)
maeArray = np.empty(46)
maeArray[:] = np.nan
maeArray[0] = mae
mseArray = np.empty(46)
mseArray[:] = np.nan
mseArray[0] = mse

df = pd.DataFrame({
    "X":axisX,
    "Y":axisY,
    "Y_hat":randomY,
    "MAE":maeArray,
    "MSE":mseArray
})
print(df)
df.to_csv('out1.csv', encoding = 'utf-8-sig')
files.download('out1.csv')
```
