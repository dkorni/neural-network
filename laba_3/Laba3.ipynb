Генерація датасету рергресії

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot
from sklearn.datasets import make_regression
from sklearn.datasets import make_classification
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

n = 100000*(30-10)

print(n)
x,y = make_regression(n_samples=n, n_features=1, noise=10)
pyplot.scatter(x,y)
pyplot.show()
```

```python
from sklearn import metrics

# Тепер, коли у нас є наші атрибути та мітки, наступним кроком є розділення цих 
# даних на навчальні та тестові набори. Ми зробимо це за допомогою 
# вбудованого методу train_test_split() Scikit-Learn
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# Тренування алгоритму
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Прогназування
y_pred = regressor.predict(X_test)

pyplot.scatter(X_test,y_pred)
pyplot.show()
pyplot.scatter(X_test,y_test)
pyplot.show()

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
# Таблиця очікуваних даних та спрогназованих
df = pd.DataFrame({'Actual': y_test[:10], 'Predicted': y_pred[:10]})
df
```

```python

X, y = make_classification(
    n_features=2, n_redundant=0, n_informative=1, n_clusters_per_class=1, n_samples = n
)

fig, ax = pyplot.subplots(figsize=(9, 6))
pyplot.xlabel("X0", fontsize=20)
pyplot.ylabel("X1", fontsize=20)
pyplot.scatter(X[:,0], X[:,1], s=50, c=y)
pyplot.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

model = KNeighborsClassifier(weights='distance', algorithm="auto", n_neighbors=5,p=2).fit(X_train, y_train)

y_pred = model.predict(X_test)

df = pd.DataFrame({'Actual': y_test[:10], 'Predicted': y_pred[:10]})
print(df)

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
